```python
import folium
import numpy as np
```

```python
f = folium.Figure(width=300, height=580)
m = folium.Map(location = [64.5, 25.748152], tiles='openstreetmap', zoom_start=5,
               attributionControl=True, zoom_control=False).add_to(f)

coord = np.array([
 #   [60.20, 24.96],
        [59.77, 21.37],
        [62.89, 27.63],
        [61.84, 24.29],
        [67.37, 26.62]])

lab_coord = np.array([
    #[60.55, 24.96],
        [60.22, 21.37],
        [63.30, 27.63],
        [62.19, 24.29],
        [67.72, 26.62]])

#lab = ['Helsinki', 'Uto', 'Vehmasmaki', 'Hyytiala', 'Sodankyla']
lab = ['Uto', 'Vehmasmaki', 'Hyytiala', 'Sodankyla']
for c, l, l_c in zip(coord, lab, lab_coord):
    folium.Marker(c.tolist(), icon=folium.Icon(color='gray', icon='rss', prefix='fa')).add_to(m)

    div = folium.DivIcon(icon_anchor = (10, -10),
                         html=(
        #'<svg height="5" width="5">'
        '<text x="0" y="0" fill="red"><h5><b>'+ l + '</text>'
        #'</svg>'
    ))
    folium.Marker(l_c.tolist(), icon=div).add_to(m)

m
```
