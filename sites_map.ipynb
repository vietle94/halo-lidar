```python
import folium
import numpy as np
```

```python
m = folium.Map(location = [65.3, 25.748152], tiles='cartodbpositronnolabels', zoom_start=5,
               attributionControl=False)

coord = np.array([[60.20, 24.96],
        [59.77, 21.37],
        [62.89, 27.63],
        [61.84, 24.29],
        [67.37, 26.62]])

lab_coord = np.array([[60.55, 24.96],
        [60.22, 21.37],
        [63.30, 27.63],
        [62.19, 24.29],
        [67.72, 26.62]])

lab = ['Helsinki', 'Uto', 'Vehmasmaki', 'Hyytiala', 'Sodankyla']
for c, l, l_c in zip(coord, lab, lab_coord):
    folium.Marker(c, icon=folium.Icon(color='gray', icon='rss', prefix='fa')).add_to(m)

    div = folium.DivIcon(html=(
        '<svg height="0" width="0">'
        '<text x="10" y="0" fill="black"><h4><b>'+ l + '</text>'
        '</svg>'
    ))
    folium.Marker(l_c, icon=div).add_to(m)

m
```
